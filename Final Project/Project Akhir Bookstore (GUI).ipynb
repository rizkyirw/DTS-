```python
from tkinter import *
from openpyxl import Workbook #--> Library for access ms. excel
from openpyxl.styles import Font,Alignment,Border,Side # --> Styling ms. excell
from tkinter import font as tkfont

#Create Object
root = Tk()
root.title("IRW Bookstore")
root.resizable(width=False,height=False)
workbook = Workbook()
sheet = workbook.active

#Create Styling font
styling = tkfont.Font(family='Helvetica',weight='bold', size=15)
styling2 = tkfont.Font(family='Helvetica', size=9)

font = Font(bold=True)
border = Border(left=Side(border_style='thin',color='00000000'),
                right=Side(border_style='thin',color='00000000'),
                top=Side(border_style='thin',color='00000000'),
                bottom=Side(border_style='thin',color='00000000'))

alignment = Alignment(horizontal='center', vertical='center')


#Create Windows 
HEIGHT = 550
WIDTH = 600
canvas = Canvas(root, height=HEIGHT, width=WIDTH, bg='lightblue')
canvas.pack()

#Create Excell Column's
sheet['A1'] = "Genre\t:"
A1 = sheet['A1']
A1.font = font
sheet['A2'] = "Tanggal Peminjaman\t:"
A2 = sheet['A2']
A2.font = font

sheet['A3'] = "No"
A3 = sheet['A3']
A3.font = font
A3.border = border
A3.alignment = alignment

sheet['B3'] = "Kode Member"
B3 = sheet['B3']
B3.font = font
B3.border = border
B3.alignment = alignment

sheet['C3'] = "Judul Buku"
C3 = sheet['C3']
C3.font = font
C3.border = border
C3.alignment = alignment

sheet['D3'] = "Kode Buku"
D3 = sheet['D3']
D3.font = font
D3.border = border
D3.alignment = alignment

sheet['E3'] = "Durasi Pinjam (Hari)"
E3 = sheet['E3']
E3.font = font
E3.border = border
E3.alignment = alignment

num = 0

#Function
def InsertData():
    global num
    num = num + 1
    sheetnum = num + 3

    sheet['A'+str(sheetnum)] = num
    DataNo = sheet['A'+str(sheetnum)]
    DataNo.border = border
    DataNo.alignment = alignment

    sheet['B'+str(sheetnum)] = memberEntry.get()
    DataMember = sheet['B'+str(sheetnum)]
    DataMember.border = border
    DataMember.alignment = alignment

    sheet['C'+str(sheetnum)] = judulEntry.get()
    DataJudul = sheet['C'+str(sheetnum)]
    DataJudul.border = border
    DataJudul.alignment = alignment

    sheet['D' + str(sheetnum)] = kodeEntry.get()
    DataKode = sheet['D' + str(sheetnum)]
    DataKode.border = border
    DataKode.alignment = alignment
    
    sheet['E' + str(sheetnum)] = durasiEntry.get()
    DataDurasi = sheet['E' + str(sheetnum)]
    DataDurasi.border = border
    DataDurasi.alignment = alignment

    sheet['B1'] = bookEntry.get()
    sheet['B2'] = tanggalEntry.get()

    memberEntry.delete(0, END)
    judulEntry.delete(0, END)
    kodeEntry.delete(0, END)
    durasiEntry.delete(0, END)

def SaveData():
    global informasi
    workbook.save(filename=str(bookEntry.get())+"_"+str(tanggalEntry.get())+".xlsx")
    informasi['text'] = "Data pinjam telah di save!\nNama file: "+str(bookEntry.get())+"_"+str(tanggalEntry.get())+".xlsx"

def CreateNewData():
    global informasi, num
    informasi['text'] = 'Klik Insert untuk menambahkan list pinjam, lalu klik Save jika semua sudah selesai.'
    memberEntry.delete(0, END)
    judulEntry.delete(0, END)
    kodeEntry.delete(0, END)
    durasiEntry.delete(0, END)
    bookEntry.delete(0, END)
    tanggalEntry.delete(0, END)
    num = 0
    
def QuitApp():
    root.destroy()

#Create Frame Input
frameTitle = Frame(root, bg='white')
frameTitle.place(rely=0.025,relx=0.5,relheight=0.1,relwidth=0.8,anchor='n')
title = Label(frameTitle, bg='white', text='IRW Book Corner', font=styling)
title.place(relheight=1,relwidth=1)

frameBook = Frame(root, bg='white')
frameBook.place(rely=0.15,relx=0.5,relheight=0.06,relwidth=0.8,anchor='n')
bookinfo = Label(frameBook, bg='white', text='Genre', font=styling2)
bookinfo.place(relwidth=0.4,relheight=1)
bookEntry = Entry(frameBook)
bookEntry.place(relx=0.4,relheight=1,relwidth=0.6)
bookEntry.get()

frameTanggal = Frame(root, bg='white')
frameTanggal.place(rely=0.22,relx=0.5,relheight=0.06,relwidth=0.8,anchor='n')
tanggalinfo = Label(frameTanggal, bg='white', text='Tanggal Peminjaman', font=styling2)
tanggalinfo.place(relwidth=0.4,relheight=1)
tanggalEntry = Entry(frameTanggal)
tanggalEntry.place(relx=0.4,relheight=1,relwidth=0.6)
tanggalEntry.get()

frameMember = Frame(root, bg='white')
frameMember.place(rely=0.29,relx=0.5,relheight=0.06,relwidth=0.8,anchor='n')
memberinfo = Label(frameMember, bg='white', text='Kode Member', font=styling2)
memberinfo.place(relwidth=0.4,relheight=1)
memberEntry = Entry(frameMember)
memberEntry.place(relx=0.4,relheight=1,relwidth=0.6)
memberEntry.get()

frameJudul = Frame(root, bg='white')
frameJudul.place(rely=0.36,relx=0.5,relheight=0.06,relwidth=0.8,anchor='n')
judulinfo = Label(frameJudul, bg='white', text='Judul Buku', font=styling2)
judulinfo.place(relwidth=0.4,relheight=1)
judulEntry = Entry(frameJudul)
judulEntry.place(relx=0.4,relheight=1,relwidth=0.6)
judulEntry.get()

frameKode = Frame(root, bg='white')
frameKode.place(rely=0.43,relx=0.5,relheight=0.06,relwidth=0.8,anchor='n')
kodeinfo = Label(frameKode, bg='white', text='Kode Buku', font=styling2)
kodeinfo.place(relwidth=0.4,relheight=1)
kodeEntry = Entry(frameKode)
kodeEntry.place(relx=0.4,relheight=1,relwidth=0.6)
kodeEntry.get()

frameDurasi = Frame(root, bg='white')
frameDurasi.place(rely=0.50,relx=0.5,relheight=0.06,relwidth=0.8,anchor='n')
durasiinfo = Label(frameDurasi, bg='white', text='Durasi Pinjam (Hari)', font=styling2)
durasiinfo.place(relwidth=0.4,relheight=1)
durasiEntry = Entry(frameDurasi)
durasiEntry.place(relx=0.4,relheight=1,relwidth=0.6)
durasiEntry.get()

#Create Box Info
informasi = Label(root, bg='white', font=styling2, text='Klik Insert untuk menambahkan list pinjam, lalu klik Save jika semua sudah selesai.')
informasi.place(rely=0.58,relx=0.5,relheight=0.1,relwidth=0.8,anchor='n')

#Create Button
frameButton = Frame(root, bg='white')
frameButton.place(rely=0.69,relx=0.5,relheight=0.3,relwidth=0.3,anchor='n')
insert = Button(frameButton, text='Insert', command=InsertData, activebackground='#345',activeforeground='white')
insert.place(rely=0,relx=0.5,relheight=0.25,relwidth=1,anchor='n')
save = Button(frameButton, text='Save', command=SaveData, activebackground='#345',activeforeground='white')
save.place(rely=0.25,relx=0.5,relheight=0.25,relwidth=1,anchor='n')
createNewData = Button(frameButton, text='Create New', command=CreateNewData, activebackground='#345',activeforeground='white')
createNewData.place(rely=0.5,relx=0.5,relheight=0.25,relwidth=1,anchor='n')
Exit = Button(frameButton, text='Exit', command=QuitApp, activebackground='red',activeforeground='white')
Exit.place(rely=0.75,relx=0.5,relheight=0.25,relwidth=1,anchor='n')





root.mainloop()
```
